```python
#importing pandas libaries and sklearn datasets
import pandas as pd
from sklearn import datasets
import numpy as np
```

```python
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data)

#Creating feature columns 
column_names = iris.feature_names
iris_df.columns = column_names
```

```python
#SPEARMAN'S. RANK CORRELATION.
iris_df.corr(method ='spearman') 
```

```python
#Importing matplotlib libary
from matplotlib import pyplot as plt
```

<h1>Iris Data Visualization<h1>

```python
plt.scatter(iris_df['sepal length (cm)'], iris.target)
plt.xlabel('sepal length (cm)')
plt.ylabel('target')
plt.show()
```

<h4>This plot shows the scatter plot of the target of the iris dataset against the sepal length (cm) feature of the data <h4>

```python
plt.scatter(iris_df['sepal width (cm)'], iris.target)
plt.xlabel('sepal width (cm)')
plt.ylabel('target')
plt.show()
```

```python
#Normalizing the iris dataset using Zscore
def z_score(df):
    df.columns = [x + "_zscore" for x in df.columns.tolist()]
    return ((df - df.mean())/df.std(ddof=0))

print(z_score(iris_df))
```
